```python
import yfinance as yf
import numpy as np
import pandas as pd
from pandas.tseries.offsets import MonthEnd
import math
```

```python
stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[2].Symbol
next_stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_Next_50')[3].Symbol
```

```python
stocks.head(5)
#next_stocks.head(5)
```

```python
stocks = stocks + '.NS'
next_stocks = next_stocks + '.NS'
```

```python
stocks = stocks.to_list()
next_stocks = next_stocks.to_list()

stocks = stocks+next_stocks
```

```python
df = yf.download(stocks,start='2023-06-19',end='2023-07-30')['Close']
```

```python
df
```

```python
ret_df = df.pct_change()
```

```python
ret_df
```

```python
ret_df.index = pd.DatetimeIndex(ret_df.index)
```

```python
mtl_ret = (ret_df + 1).resample('M').prod()
```

```python
mtl_ret
```

## Momentum Calculation

```python
returns_over_lookback = mtl_ret.rolling(2).apply(np.prod).dropna()
```

```python
returns_over_lookback
```

```python
#we vary the argument of the nlargest function from 5 to 25 in steps of 5 to obtain our data

def top_performers(date,n):
    all_ = returns_over_lookback.loc[date]
    top = all_.nlargest(n)
    return top

def bottom_performers(date,n):
    all_ = returns_over_lookback.loc[date]
    top = all_.nsmallest(n)
    return top

print(top_performers('2023-07-31',10))
print(bottom_performers('2023-07-31',10))
```
